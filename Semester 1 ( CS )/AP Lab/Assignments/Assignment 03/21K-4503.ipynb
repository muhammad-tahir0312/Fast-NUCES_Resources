### QUESTION :02

###### Write a function Projectile () such that it can accept two variables initial velocity and angle and calculate the Maximum Height and Range of projectile. And also it must return both Maximum Height and Range in a single return call. Also plot the graph of Maximum Height and Range with respect to angle from 0 to 360 degree using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

def projectile(vi,angle):
    r = ((vi**2)*(np.sin(angle*angle)))/(2*9.8)
    h = ((vi**2)*(np.sin(2*angle)))/(9.8)
    a = np.arange(0.0, 360.0, 0.1)
        
    plt.plot(a, ((vi**2)*(np.sin(np.radians(a))**2))/(2*9.8), color='red', linestyle='dashed', linewidth = 3, marker='o', markerfacecolor='blue', markersize=12)   
    plt.xlabel('Acceleration')
    plt.ylabel("Range")
    plt.title('Acceleration VS Range') 
    plt.show()
    
    plt.plot(a, ((vi**2)*2*np.sin(np.radians(a)))/9.8, color='red', linestyle='dashed', linewidth = 3, marker='o', markerfacecolor='blue', markersize=12)
    plt.xlabel('Acceleration')
    plt.ylabel('Height')
    plt.title('Acceleration VS Height') 
    plt.show()
    
    return r , h
```

```python
print("The Maximum Range and Height of Projectile is: " + str(Projectile(5,45)))
```

### QUESTION :03

###### Write a function SHM() such that it can accept four variables (amplitude , angular frequency , phase constant , time ) and return displacement , velocity ,and acceleration in a single call.
###### x = xm Cos (t + )
###### v = - xm  Sin (t + )
###### a = xm  2 Cos (t + )

```python
import numpy as np
import matplotlib.pyplot as plt

def SHM(amp, ang, phase, time):
  x = amp*np.cos(np.radians(ang*time + phase))
  v = -amp*ang*np.sin(np.radians(ang*time + phase))
  a = amp*ang**2*np.sin(np.radians(ang*time + phase))

  t = np.arange(0.0, 10.0, 0.1)
  plt.plot(t, amp*np.cos(np.radians(ang*t + phase)), color='green', linestyle='dashed', linewidth = 5, marker='o', markerfacecolor='orange', markersize=4)
  plt.xlabel('Time')
  plt.ylabel('Displacement')
  plt.title('Time VS Displacement') 
  plt.show()
    
  plt.plot(t, -amp*ang*np.sin(np.radians(ang*t + phase)), color='green', linestyle='dashed', linewidth = 5, marker='o', markerfacecolor='orange', markersize=4)
  plt.xlabel('Time')
  plt.ylabel('Velocity')
  plt.title('Time VS Velocity') 
  plt.show()
    
  plt.plot(t, amp*ang**2*np.sin(np.radians(ang*t + phase)), color='green', linestyle='dashed', linewidth = 5, marker='o', markerfacecolor='orange', markersize=4)
  plt.xlabel('Time')
  plt.ylabel('Acceleration')
  plt.title('Time VS Acceleration')
  plt.show()

  return x, v, a
```

```python
print(SHM(10,20,30,40))
```
